```python
import tweepy
from textblob import TextBlob #NLP bahasa inggris
import re
import pandas as pd
```

```python
data = pd.read_csv('Dataset/All-Pra & Pasca ND Clean for translate 5.csv',sep=";")#nama file data
data = data.astype({'Tweet' : 'string'})
data = data.astype({'Terjemah' : 'string'})
data.dtypes
```

```python
def load_data():
    data = pd.read_csv('Dataset/All-Pra & Pasca ND Clean for translate 5.csv',sep=";")#nama file data
    return data
```

```python
tweet_df = load_data()
tweet_df.head()
```

```python
df  = pd.DataFrame(tweet_df[['Tanggal', 'Banyak Retweet', 'Tweet', 'Terjemah']])
df
```

```python
def getSubjectivity(text): 
    return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity

df['Subjectivity'] = df['Terjemah'].apply(getSubjectivity)
df['Polarity'] = df['Terjemah'].apply(getPolarity)

df
```

```python
def getSentiment(score):
    if score < 0 :
        return '0'
    elif score == 0 :
        return '1'
    else :
        return '2'
# Apply the function
df['Sentiment'] = df['Polarity'].apply(getSentiment)
#lets take a look
df
```

```python
for index, row in df.iterrows():
  id_blob = df.iloc[index]['Tweets']
  translation = TextBlob(id_blob)
  if translation.detect_language() != 'en' :
    en_blob = translation.translate(to=u'en')
  else :
    en_blob = translation
  df.at[index,str('Tweets_Translated')] = str(en_blob)
# Lets take a look
df.head()
```
